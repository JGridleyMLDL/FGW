```python
g1=Graph()
g1.add_attibutes({0:1,1:7,2:1,3:7})
g1.add_edge((0,1))
g1.add_edge((1,2))
g1.add_edge((3,0))
g1.add_edge((3,2))
g1.add_edge((0,2))
g2=Graph()
g2.add_attibutes({0:7,1:1,2:7,3:1})
g2.add_edge((0,1))
g2.add_edge((1,2))
g2.add_edge((3,0))
g2.add_edge((3,2))
g2.add_edge((2,0))
```

```python
plt.figure(figsize=(5,4))
vmin=0
vmax=7
draw_rel(g1.nx_graph,draw=False,vmin=vmin,vmax=vmax,with_labels=False)
draw_rel(g2.nx_graph,draw=False,vmin=vmin,vmax=vmax,with_labels=False,shiftx=5,swipy=True)
plt.show()
```

```python
alpha=0.5
dfgw=Fused_Gromov_Wasserstein_distance(alpha=alpha,features_metric='dirac',method='shortest_path').graph_d(g1,g2)
dw=Wasserstein_distance(features_metric='dirac').graph_d(g1,g2)
dgw=Fused_Gromov_Wasserstein_distance(alpha=1,features_metric='dirac',method='shortest_path').graph_d(g1,g2)
print('Wasserstein distance={}, Gromov distance={} \nFused Gromov-Wasserstein distance for alpha {} = {}'.format(dw,dgw,alpha,dfgw))
```

```python
def build_comunity_graph(N=30,mu=0,sigma=0.3,pw=0.8):
    v=mu+sigma*np.random.randn(N);
    g=Graph()
    g.add_nodes(list(range(N)))
    for i in range(N):
         g.add_one_attribute(i,v[i])
         for j in range(N):
             r=np.random.rand()
             if  r<pw:
                 g.add_edge((i,j))
    return g
```

```python
N=13
Nc=2
np.random.seed(12)
g1=build_comunity_graph(N=N,mu=10,sigma=0.2,pw=0.5)
g2=build_comunity_graph(N=N,mu=0,sigma=0.2,pw=0.5)
```

```python
plt.figure(figsize=(10,5))
draw_rel(g1.nx_graph,vmin=-1,vmax=1,with_labels=False,draw=False)
draw_rel(g2.nx_graph,vmin=-1,vmax=1,with_labels=False,shiftx=3,draw=False)
plt.show()
```

```python
def merge_graph(g1,g2):
    gprime=nx.Graph(g1)
    N0=len(gprime.nodes())
    g2relabel=nx.relabel_nodes(g2, lambda x: x +N0)
    gprime.add_nodes_from(g2relabel.nodes(data=True))
    gprime.add_edges_from(g2relabel.edges(data=True)) 
    gprime.add_edge(N0-1,N0)
    
    return gprime
```

```python
g3=merge_graph(g1.nx_graph,g2.nx_graph)
```

```python
plt.figure(figsize=(8,5))
draw_rel(g3,vmin=-1,vmax=1,with_labels=False,draw=False)
plt.show()
```

```python
temp=build_comunity_graph(N=N,mu=10,sigma=0.2,pw=0.5).nx_graph
g4=merge_graph(g3,temp,k=10)
temp=build_comunity_graph(N=N,mu=0,sigma=0.2,pw=0.5).nx_graph
g5=merge_graph(g3,temp,k=20)
```

```python
plt.figure(figsize=(8,5))
draw_rel(g4,vmin=-1,vmax=1,with_labels=False,draw=False)
draw_rel(g5,vmin=-1,vmax=1,with_labels=False,shiftx=3,draw=False)
plt.show()
```
